```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
customer_df = pd.read_csv("files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
customer_df
```

```python
# Display the first few rows
print(customer_df.head())
```

```python
# Check the shape of the dataset
print(customer_df.shape)
```

```python
# Get information about the columns and data types
print(customer_df.info())
```

```python
# Rename columns to snake case
customer_df.columns = customer_df.columns.str.lower().str.replace(" ", "_")
```

```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df.dtypes
```

```python
# Plot correlation matrix
correlation_matrix = customer_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Plot continuous variables
continuous_vars = ["customer_lifetime_value", "income", "monthly_premium_auto", "months_since_last_claim",
                   "months_since_policy_inception", "total_claim_amount"]

for column in continuous_vars:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=customer_df, x=column, bins=30, kde=True)
    plt.title(f"Distribution of {column}")
    plt.show()
```

```python
# Delete customer_id column
del customer_df["customer"]

# Plot categorical variables
categorical_vars = customer_df.select_dtypes(include="object").columns

for column in categorical_vars:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=customer_df, x=column)
    plt.title(f"Count of {column}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Check for outliers in continuous variables
numerical_columns = customer_df.select_dtypes("number")
for col in numerical_columns:
    sns.boxplot(data=customer_df, y=col)
    plt.title(col)
    plt.show()
```

```python
# Check for missing values
missing_values = customer_df.isnull().sum()
print(missing_values)
```
